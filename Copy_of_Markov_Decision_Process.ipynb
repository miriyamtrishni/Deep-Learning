```python
import copy
import timeit
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline
from ipywidgets import interactive
import ipywidgets as widgets
```

Create the following grid world.

**States:** A 10x10 grid

**Actions:** Up, Down, Left, Right

**Tranistion probabilities:**
* 0.7 in the direction of action
* 0.1 in the three other directions
* The robot bounces back to the same state near edges

**Rewards:**
* (7,8) has a reward +10
* (2,7) has a reward +3
* (4,3) has a reward -5
* (7,3) has a reward -10
* No reward in other states

This example is based on Decision Making Under Uncertainty by M.J. Kochenderfer.

```python
#Let's define MDP paras
def createGrid10World():
    def xy2s(y, x):
        x = max(x, 0)
        y = max(y, 0)
        x = min(x, 9)
        y = min(y, 9)
        out = np.ravel_multi_index(np.array([x,y]), (10,10))
        return out

    def s2xy(s):
        x, y = np.unravel_index(s, (10,10))
        return y, x

    def gridPlot(ax, im, title='', cmap='Blues'):
        pl.imshow(im, interpolation='none', cmap=cmap, origin='lower')
        pl.colorbar()
        ax.set_xticks(np.arange(0, 10, 1));
        ax.set_yticks(np.arange(0, 10, 1));
        ax.set_xticklabels(np.arange(0, 10, 1));
        ax.set_yticklabels(np.arange(0, 10, 1));
        ax.set_xticks(np.arange(-.5, 10, 1), minor=True);
        ax.set_yticks(np.arange(-.5, 10, 1), minor=True);
        ax.grid(which='minor', color='w', linestyle='-', linewidth=1)
        pl.title(title);
        return

    A = ['left', 'right', 'up', 'down']
    S = np.arange(100)
    T = np.zeros((len(S), len(A), len(S)))
    R = np.zeros((len(S), len(A)))
    for s in S:
        x, y = s2xy(s)
        if x == 2 and y == 7:
            R[s, :] = 3
        elif x == 7 and y == 8:
            R[s, :] = 10
        else:
            if x == 7 and y == 3:
                R[s, :] = -10
            elif x == 4 and y == 3:
                R[s, :] = -5
            elif x == 0:
                if y == 0 or y == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 0] = -0.7
            elif x == 9:
                if y == 0 or y == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 1] = -0.7
            elif y == 0:
                if x == 0 or x == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 3] = -0.7
            elif y == 9:
                if x == 0 or x == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 2] = -0.7

            for a in A:
                if a == 'left':
                    T[s, 0, xy2s(x - 1, y)] += 0.7
                    T[s, 0, xy2s(x + 1, y)] += 0.1
                    T[s, 0, xy2s(x, y - 1)] += 0.1
                    T[s, 0, xy2s(x, y + 1)] += 0.1
                elif a == 'right':
                    T[s, 1, xy2s(x + 1, y)] += 0.7
                    T[s, 1, xy2s(x - 1, y)] += 0.1
                    T[s, 1, xy2s(x, y - 1)] += 0.1
                    T[s, 1, xy2s(x, y + 1)] += 0.1
                elif a == 'up':
                    T[s, 2, xy2s(x, y + 1)] += 0.7
                    T[s, 2, xy2s(x, y - 1)] += 0.1
                    T[s, 2, xy2s(x - 1, y)] += 0.1
                    T[s, 2, xy2s(x + 1, y)] += 0.1
                elif a == 'down':
                    T[s, 3, xy2s(x, y - 1)] += 0.7
                    T[s, 3, xy2s(x, y + 1)] += 0.1
                    T[s, 3, xy2s(x - 1, y)] += 0.1
                    T[s, 3, xy2s(x + 1, y)] += 0.1

    for a, c_x, c_y in [(0,0,0), (0,0,9),(1,9,0),(1,9,9),(2,0,9),(2,9,9),(3,0,0),(3,9,0)]:
        R[xy2s(c_x,c_y),a] = -0.8

    discount = 0.9

    nextStates = {}
    for si in range(len(S)):
        for ai in range(len(A)):
            nextStates[(si,ai)] = np.where((T[si, ai, :] != 0) == True)[0]

    return {'S':S, 'A':A, 'T':T, 'R':R, 'discount':discount, 'nextStates':nextStates, 'gridPlot':gridPlot, 'xy2s':xy2s, 's2xy':s2xy}
```

```python
class MDP():
    def __init__(self):
        pass

    def inbuilt_init(self, mdp_dict):
        self.S = mdp_dict['S']
        self.A = mdp_dict['A']
        self.T = mdp_dict['T']
        self.R = mdp_dict['R']
        self.discount = mdp_dict['discount']
        self.nextStates = mdp_dict['nextStates']
        self.gridPlot = mdp_dict['gridPlot']
        self.xy2s = mdp_dict['xy2s']
        self.s2xy = mdp_dict['s2xy']
```

```python
#Define the MDP
mdp = MDP()
mdp.inbuilt_init(mdp_dict=createGrid10World())

#Plot states
pl.figure(figsize=(3,3))
mdp.gridPlot(ax=pl.gca(), im=mdp.S.reshape((10,10)), title='States', cmap='Greys')

#Plot rewards
pl.figure(figsize=(15,3))
pl.suptitle('Rewards', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=mdp.R[:,a].reshape((10,10)), title='a='+mdp.A[a], cmap='jet')
pl.show()

#Plot rewards
pl.figure(figsize=(15,3))
pl.suptitle('Rewards - clipped larger +ve values for visualizing the edges', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=np.clip(mdp.R[:,a].reshape((10,10)), -1, -0.5), title='a='+mdp.A[a], cmap='jet')
pl.show()

#Plot rewards
s0_x, s0_y = 3, 5
s0 = mdp.xy2s(s0_x, s0_y)
pl.figure(figsize=(15,3))
pl.suptitle('Transition probabilities T(s1|s0=({},{}),a)'.format(s0_x, s0_y), fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=mdp.T[s0,a,:].reshape((10,10)), title='a='+mdp.A[a], cmap='Blues')
pl.show()
```

### 1. Policy evaluation

Computing the utility, U.

$U^\pi_k(s) = R(s, \pi(s)) + \gamma \sum_{s'} T(s' \mid s, \pi(s))U^\pi_{k-1}(s')$

```python
def iterativePolicyEvaluation(mdp, policy, numIterations=10):
    """
    Performs iterative policy evaluation for a (possibly) fixed policy.
    `policy` can be either:
      - an integer action index (same action for all states), or
      - an array/list of action indices of length |S| (action per state).
    Returns U (numpy array of length |S|).
    """
    U = np.zeros(len(mdp.S))
    for t in range(numIterations):
        U_old = U.copy()
        for s in range(len(mdp.S)):
            # resolve action for this state (support scalar or per-state policy)
            if isinstance(policy, (list, tuple)) or (hasattr(policy, 'shape') and len(policy)==len(mdp.S)):
                a = int(policy[s])
            else:
                a = int(policy)
            # expected utility under policy
            total = 0.0
            for s2 in mdp.nextStates[(s,a)]:
                total += mdp.T[s, a, s2] * U_old[s2]
            U[s] = mdp.R[s, a] + mdp.discount * total
    return U


numIterations = 5
pl.figure(figsize=(15,3))
pl.suptitle('Utilities', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    U = iterativePolicyEvaluation(mdp=mdp, policy=a, numIterations=numIterations)
    mdp.gridPlot(ax=pl.gca(), im=U.reshape(10,10), title='a='+mdp.A[a], cmap='jet')
pl.show()
#print(np.round(U.reshape(10,10),3))
```

```python
#Value iteration
def valueIteration(mdp, numIterations=1):
    """Performs value iteration for `numIterations` iterations and returns the value function U."""
    U = np.zeros(len(mdp.S))
    for t in range(numIterations):
        U_old = U.copy()
        for s in range(len(mdp.S)):
            action_values = []
            for a in range(len(mdp.A)):
                total = 0.0
                for s2 in mdp.nextStates[(s,a)]:
                    total += mdp.T[s, a, s2] * U_old[s2]
                action_values.append(mdp.R[s,a] + mdp.discount * total)
            U[s] = max(action_values)
    return U

def policyExtration(mdp, U):
    """Extracts a greedy policy from a given value function U."""
    policy = np.zeros(len(mdp.S), dtype=int)
    for s in range(len(mdp.S)):
        best_a = 0
        best_val = -1e9
        for a in range(len(mdp.A)):
            total = 0.0
            for s2 in mdp.nextStates[(s,a)]:
                total += mdp.T[s,a,s2] * U[s2]
            val = mdp.R[s,a] + mdp.discount * total
            if val > best_val:
                best_val = val
                best_a = a
        policy[s] = best_a
    return policy

U = valueIteration(mdp, numIterations=2)
policy = policyExtration(mdp, U=U)
pl.figure(figsize=(3,3))
mdp.gridPlot(ax=pl.gca(), im=U.reshape(10,10), title='Utility', cmap='jet')
for s in range(100):
    x, y = mdp.s2xy(s)
    if policy[s] == 0:
        m='\u02C2'
    elif policy[s] == 1:
        m='\u02C3'
    elif policy[s] == 2:
        m='\u02C4'
    elif policy[s] == 3:
        m='\u02C5'
    pl.text(x-0.5,y-1,m,color='k',size=20)
pl.show()
```

### 2. Policy iteration

Policy evaluation can be used in policy iteration:
1. Given the current policy, compute U
2. Using U, compute a new policy

```python
def policyIteration(mdp, numIterations=1):
    """Standard policy iteration:
    - Start with a random policy pi_k
    - Evaluate pi_k (iteratively until convergence)
    - Improve policy greedily
    Repeat until policy is stable or numIterations reached.
    Returns (U_pi_k, pi_kp1)
    """
    U_pi_k = np.zeros(len(mdp.S)) #initial values
    pi_k = np.random.randint(low=0,high=len(mdp.A),size=len(mdp.S),dtype=int) #initial policy
    pi_kp1 = copy.copy(pi_k)
    for t in range(numIterations):
        #Policy evaluation: compute U_pi_k (iterative evaluation until convergence)
        theta = 1e-6
        U = U_pi_k.copy()
        while True:
            delta = 0.0
            U_old = U.copy()
            for s in range(len(mdp.S)):
                a = int(pi_k[s])
                total = 0.0
                for s2 in mdp.nextStates[(s,a)]:
                    total += mdp.T[s,a,s2] * U_old[s2]
                U[s] = mdp.R[s,a] + mdp.discount * total
                delta = max(delta, abs(U[s]-U_old[s]))
            if delta < theta:
                break
        U_pi_k = U
        #Policy improvement
        stable = True
        for s in range(len(mdp.S)):
            best_a = None
            best_val = -1e9
            for a in range(len(mdp.A)):
                total = 0.0
                for s2 in mdp.nextStates[(s,a)]:
                    total += mdp.T[s,a,s2] * U_pi_k[s2]
                val = mdp.R[s,a] + mdp.discount * total
                if val > best_val:
                    best_val = val
                    best_a = a
            pi_kp1[s] = int(best_a)
            if pi_kp1[s] != pi_k[s]:
                stable = False
        if stable:
            break
        pi_k = pi_kp1.copy()
    return U_pi_k, pi_kp1

U_pi_k, pi_kp1 = policyIteration(mdp, numIterations=2)
```

```python
import time
def value_iteration_until_converged(mdp, theta=1e-6):
    U = np.zeros(len(mdp.S))
    start = time.perf_counter()
    while True:
        delta = 0.0
        U_old = U.copy()
        for s in range(len(mdp.S)):
            action_values = []
            for a in range(len(mdp.A)):
                total = 0.0
                for s2 in mdp.nextStates[(s,a)]:
                    total += mdp.T[s,a,s2] * U_old[s2]
                action_values.append(mdp.R[s,a] + mdp.discount * total)
            U[s] = max(action_values)
            delta = max(delta, abs(U[s] - U_old[s]))
        if delta < theta:
            break
    end = time.perf_counter()
    return U, end - start

U_vi, t_vi = value_iteration_until_converged(mdp)
print("Value iteration time (s):", t_vi)
```

```python
import numpy as np, time

def q_learning_on_mdp(mdp, episodes=2000, alpha=0.1, gamma=None, epsilon=0.1, max_steps=200):
    if gamma is None:
        gamma = mdp.discount
    Q = np.zeros((len(mdp.S), len(mdp.A)))
    rewards = []
    start = time.perf_counter()
    for ep in range(episodes):
        s = np.random.randint(len(mdp.S))
        cum_r = 0.0
        for step in range(max_steps):
            # epsilon-greedy
            if np.random.rand() < epsilon:
                a = np.random.randint(len(mdp.A))
            else:
                a = int(np.argmax(Q[s,:]))
            # sample next state safely (normalize probs)
            probs = mdp.T[s,a,:]
            if probs.sum() == 0:
                # If no transitions defined, stay in same state
                s_prime = s
            else:
                s_prime = np.random.choice(len(mdp.S), p=probs/probs.sum())
            r = mdp.R[s,a]
            Q[s,a] = Q[s,a] + alpha * (r + gamma * np.max(Q[s_prime,:]) - Q[s,a])
            cum_r += r
            s = s_prime
        rewards.append(cum_r)
    end = time.perf_counter()
    V = Q.max(axis=1)
    policy = Q.argmax(axis=1)
    return Q, V, policy, rewards, end-start
```

```python
Q, V_q, pi_q, rewards, t_q = q_learning_on_mdp(mdp, episodes=2000, alpha=0.1, epsilon=0.1)
print("Q-learning time (s):", t_q)
print("Final average episode reward:", np.mean(rewards[-100:]))  # last 100 episodes
```
