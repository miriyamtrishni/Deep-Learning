```python


plt.hist(np.random.randn(100000,1), density=True, bins=200, histtype="step", color="green", label="Normal")
plt.hist(np.random.rand(100000,1), density=True, bins=200, histtype="step", color="blue", label="Uniform")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python

# Create x and y values in range [-5, 5]
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Equation Z = X^2 + Y^2
Z = X**2 + Y**2

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=matplotlib.cm.coolwarm, linewidth=0.1)

# Labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z = X² + Y²")
plt.show()
```
