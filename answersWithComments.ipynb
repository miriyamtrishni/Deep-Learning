```python
# Generate samples for plotting histograms
normal_data = np.random.randn(100000, 1)  # Normal distribution (mean=0, std=1)
uniform_data = np.random.rand(100000, 1)  # Uniform distribution [0, 1]
```

```python
# Plot histograms for comparison
plt.hist(normal_data, density=True, bins=200, histtype="step", color="green", label="Normal")
plt.hist(uniform_data, density=True, bins=200, histtype="step", color="blue", label="Uniform")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Create X, Y coordinate grids
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
# Apply formula
Z = X**2 + Y**2
```

4. 3D PLOT OF Z = X² + Y²

```python

# Plot 3D surface
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
```

```python
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z = X² + Y²")
plt.show()
```

5. CORRELATION ANALYSIS (Online Store Dataset)

```python
store_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/online_store_customer_data.csv")

# Select numeric features
features = ['Age', 'Referal', 'Amount_spent']
data = store_df[features]
```

```python
# Calculate correlations
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')
```

```python
# Plot heatmaps
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Correlation")
```

```python
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Spearman Correlation")
```

```python
plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------------
# 6. FILTER COLUMNS & SHOW FIRST 20 SAMPLES
# ----------------------------------------------
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
print("\nFirst 20 samples of filtered dataframe:")
display(new_df.head(20))
```

```python
# ----------------------------------------------
# 7. DESCRIPTIVE STATISTICS
# ----------------------------------------------
stats_df = pd.DataFrame({
    'Mean': new_df.mean(),
    'Standard Deviation': new_df.std(),
    '25% Quantile': new_df.quantile(0.25),
    '50% Quantile (Median)': new_df.quantile(0.50),
    '75% Quantile': new_df.quantile(0.75)
})
print("\nStatistical Summary of selected features:")
display(stats_df)
```

```python
# ----------------------------------------------
# 8. HISTOGRAMS FOR EACH FEATURE
# ----------------------------------------------
plt.figure(figsize=(20, 4))
for i, col in enumerate(new_df.columns):
    plt.subplot(1, 4, i+1)
    plt.hist(new_df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------------
# 9. PAIRPLOT (Scatter plots + Histograms)
# ----------------------------------------------
sns.pairplot(new_df)
plt.show()
```

```python
# ----------------------------------------------
# 10. CUSTOM PAIRGRID (Different plots in upper, lower, diagonal)
# ----------------------------------------------
g = sns.PairGrid(new_df)

# Diagonal: histogram + KDE
g.map_diag(sns.histplot, kde=True)
```

```python
# Lower half: bivariate KDE with filled contours
g.map_lower(sns.kdeplot, fill=True)
```

```python
# Create a PairGrid for the dataframe
g = sns.PairGrid(new_df)

# Upper half: scatter plots
g.map_upper(sns.scatterplot)

# Diagonal: histogram + KDE
g.map_diag(sns.histplot, kde=True)

# Lower half: bivariate KDE
g.map_lower(sns.kdeplot, fill=True)

plt.show()
```
